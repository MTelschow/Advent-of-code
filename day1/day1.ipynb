```julia

inputs = Vector{Vector{Char}}()
# Read file
open("input.txt") do f
    while ! eof(f)  
        textLine = readline(f)
        charArray = Vector{Char}()
        for letter in textLine
            push!(charArray, letter)
        end  
        push!(inputs, charArray)
    end
   
end

println(inputs)
```

```julia
numbers = [
    "one",
    "two",
    "three",
    "four",
    "five",
    "six",
    "seven",
    "eight",
    "nine"
]

alphaNumbers = Vector{Vector{Char}}()

for numString in numbers
    alphaNumberLetters = Vector{Char}()

    for letter in numString
        push!(alphaNumberLetters, letter)
    end

    push!(alphaNumbers, alphaNumberLetters)
end
println(alphaNumbers)
```

```julia
function getSum(inputs, a=false)
    sum = 0
    
    for inputIdx in 1:length(inputs)
        arrayString = inputs[inputIdx]
        println(inputIdx, " ", getFirstDigit(arrayString, a), " ", getLastDigit(arrayString, a))
        println(inputIdx, " ", getFirstDigit(arrayString, false), " ", getLastDigit(arrayString, false))
        sum += getFirstDigit(arrayString, a) * 10 + getLastDigit(arrayString, a)
    end

    return sum
end

function getFirstDigit(charArray, a)
    for index in 1:length(charArray)
        value = checkIfNumber(index, charArray, a)
        if value != 0
            return value
        end
    end 
end

function getLastDigit(charArray, a)
    lastDigit = 0
    for index in 1:length(charArray)
        value = checkIfNumber(index, charArray, a)
        if value != 0
            lastDigit = value
        end
    end 
    return lastDigit
end

function checkIfNumber(index, charArray, a)
    # If is 1-9
    if '1' <= charArray[index] <= '9'
        return parse(Int64, charArray[index])
    end
    # If not alpha return
    if !a return 0 end

    length = size(charArray, 1)

    for alphaIdx in 1:size(alphaNumbers, 1)
        currentString = alphaNumbers[alphaIdx]

        # if length < (size(currentString, 1) + index - 1) 
        #     continue
        # end

        isNumber = true
        for i in 1:size(currentString, 1)
            if charArray[i] != currentString[i]
                isNumber = false
                break
            end
        end

        if isNumber return alphaIdx end
    end

    return 0
end

# println(getSum(inputs))
println(getSum(inputs, true))
```
